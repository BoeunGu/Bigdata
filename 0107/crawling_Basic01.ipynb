```python
import requests 
```

```python
from bs4 import BeautifulSoup
```

```python
resonse = requests.get("http://www.naver.com") # 웹페이지에 내가 원하는 정보를 가져온 것을 변수에 저장
```

```python
html=resonse.text
print(html)
```

```python
soup=BeautifulSoup(html,'html.parser') # html.parser(html 번역기)로 해석
print(soup)
word=soup.select_one('#NM_set_home_btn') # soup에서 해당하는 id 값을 가진 태크를 추출
print(word)
print(word.attrs['href']) #태그내의 속성의 값만 추출해줌
```

```python
print(word.text)
```

```python
# select는 모든 태그 선택, select_one은 처음의 한개 태그만 선택
todaynews=soup.select_one('strong.title') #Strong 태그의 클래스이름이 title인 것을 선택 
```

```python
print(todaynews.text)
```

```python
word2=soup.select('strong.title')
word2
```

```python
# for 변수 in 배열, 배열의 값을 하나씩 변수에 대입
for i in word2:
    print(i.text)
```
